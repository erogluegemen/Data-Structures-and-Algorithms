# Mükemmel İkili Ağaç
- Mükemmel ikili ağaç, her düğümün tam olarak iki alt düğüme sahip olduğu ve tüm yaprak düğümlerinin aynı seviyede olduğu bir ikili ağaç türüdür.
- Tüm düğümlerin derecesi 2'dir.
- Özyinelemeli olarak, mükemmel bir ikili ağaç şu şekilde tanımlanabilir:
    1. Tek bir düğümün çocuğu yoksa, h = 0 yüksekliğinde mükemmel bir ikili ağaçtır.
    2. Bir düğüm h > 0'a sahipse, alt ağaçlarının her ikisi de h - 1 yüksekliğindeyse ve örtüşmüyorsa, bu mükemmel bir ikili ağaçtır.



```python
### Pythonda Mükemmel İkili Ağaç İşlemi ###


class newNode:
    def __init__(self, k):  # Düğümün değerleri tanımlanır.
        self.key = k
        self.right = self.left = None


# Derinliği hesaplamak.
def calculateDepth(node):
    d = 0
    while (node is not None):
        d += 1
        node = node.left
    return d


# Mükemmel İkili Ağaç mı diye kontrol etmek.
def is_perfect(root, d, level=0):

    # Boş olma durumunu kontrol etmek.
    if (root is None):
        return True

    # Ağaçların varlığını kontrol etmek.
    if (root.left is None and root.right is None):
        return (d == level + 1)

    if (root.left is None or root.right is None):
        return False

    return (is_perfect(root.left, d, level + 1) and
            is_perfect(root.right, d, level + 1))


root = None
root = newNode(1)
root.left = newNode(2)
root.right = newNode(3)
root.left.left = newNode(4)
root.left.right = newNode(5)

if (is_perfect(root, calculateDepth(root))):
    print("Ağaç: İkili Mükemmel Ağaç.")
else:
    print("Ağaç: İkili Mükemmel Ağaç Değil.")
```
