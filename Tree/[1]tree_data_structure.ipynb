# Ağaç Veri Yapısı
- Ağaç, birbirine bağlanan düğümlerden oluşan doğrusal olmayan hiyerarşik bir veri yapısıdır.<br>


### Neden Ağaç Veri Yapısı?
- Diziler, bağlantılı liste, yığın ve kuyruk gibi diğer veri yapıları, verileri sıralı olarak depolayan <b>doğrusal veri yapılarıdır.</b> Doğrusal bir veri yapısında herhangi bir işlemi gerçekleştirmek için veri boyutunun artmasıyla zaman karmaşıklığı(time complexity'si) artar. Ancak, günümüzün hesaplama dünyasında kabul edilebilir değildir.

- Farklı ağaç veri yapıları, doğrusal olmayan bir veri yapısı olduğu için verilere daha hızlı ve daha kolay erişim sağlar.
    ### Düğüm:
    - Düğüm, bir anahtar veya değer içeren ve alt düğümlerine işaret eden bir varlıktır.
    - Her yolun son düğümlerine yaprak düğümler veya alt düğümlere bir bağlantı/işaretçi içermeyen dış düğümler denir.
    - En az bir alt düğüme sahip olan düğüme dahili düğüm denir.
    
    ### Kenar:
    - İki düğümü birbirine bağlar.<br>
    
    
    ### Kök:
    - Ağacın en üstündeki düğümdür.
    
    ### Düğümün Yüksekliği:
    - Bir düğümün yüksekliği, <b>düğümden en derin yaprağa kadar olan kenarların sayısıdır.</b><br> 
        (yani, düğümden yaprak düğümüne en uzun yol).
        
    ### Düğümün Derinliği:
    - Bir düğümün derinliği, <b>kökten düğüme kadar olan kenarların sayısıdır.</b>
    
    ### Düğümün Derecesi:
    - Bir düğümün derecesi, o düğümün <b>toplam dal sayısıdır.</b>
    
    ### Orman:
    - Birbirinden kopuk ağaçların oluşturduğu topluluğa orman denir.
    - Bir ağacın kökünü keserek bir orman oluşturulabilir.
    
    ### Ağaç Türleri:
    - İkili Ağaç
    - İkili Arama Ağacı
    - AVL Ağacı
    - B Ağacı
