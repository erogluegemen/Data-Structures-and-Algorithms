# İkili Ağaç 
- İkili ağaç, her bir üst düğümün en fazla iki çocuğa sahip olabileceği bir ağaç veri yapısıdır.<br>
  Bir ikili ağacın her düğümü üç öğeden oluşur:
  1. Veri öğesi
  2. Sol çocuğun adresi
  3. Sağ çocuğun adresi<br>
  
  

### İkili Ağaç Türleri:
1. <b>Tam İkili Ağaç:</b><br> Tam İkili ağaç, her üst düğümün/iç düğümün iki veya hiç çocuğu olmayan özel bir ikili ağaç türüdür.<br>
2. <b>Mükemmel İkili Ağaç:</b><br> Mükemmel ikili ağaç, her düğümün tam olarak iki alt düğüme sahip olduğu ve tüm yaprak düğümlerinin aynı seviyede olduğu bir ikili ağaç türüdür.<br>
3. <b>Komple İkili Ağaç:</b><br>Komple ikili ağaç, tam bir ikili ağaç gibidir, ancak iki büyük farkı vardır.
    1. Her seviye tamamen doldurulmalıdır.
    2. Tüm yaprak elemanları sola doğru eğilmelidir.
    3. Son yaprak elemanının bir doğru kardeşi olmayabilir, yani komple bir ikili ağacın tam bir ikili ağaç olması gerekmez.<br>
4. <b>Dejenere veya Patolojik Ağaç:</b><br>Dejenere veya patolojik bir ağaç, sağda veya solda tek bir çocuğu olan ağaçtır.
5. <b>Çarpık İkili Ağaç:</b><br>Çarpık ikili ağaç, ağaca sol düğümlerin veya sağ düğümlerin hakim olduğu patolojik/dejenere bir ağaçtır. <br>Böylece, iki tür çarpık ikili ağaç vardır: <b>sola çarpık ikili ağaç</b> ve <b>sağa çarpık ikili ağaç.</b>
6. <b>Dengeli İkili Ağaç:</b><br>Her düğüm için sol ve sağ alt ağacın yüksekliği arasındaki farkın 0 veya 1 olduğu bir ikili ağaç türüdür.


```python
### Pythonda İkili Ağaç İşlemi ###

class Node:
    def __init__(self, key):  # Ağacın değerleri tanımlanır.
        self.left = None
        self.right = None
        self.val = key

        
    def traversePreOrder(self):  # Ön sıralı geçiş fonksiyonu tanımlanır.
        print(self.val, end=' ')
        if self.left:
            self.left.traversePreOrder()
        if self.right:
            self.right.traversePreOrder()


    def traverseInOrder(self):  # Sıralı geçiş fonksiyonu tanımlanır.
        if self.left:
            self.left.traverseInOrder()
        print(self.val, end=' ')
        if self.right:
            self.right.traverseInOrder()

            
    def traversePostOrder(self):  # Arka sıralı geçiş fonksiyonu tanımlanır.
        if self.left:
            self.left.traversePostOrder()
        if self.right:
            self.right.traversePostOrder()
        print(self.val, end=' ')


root = Node(1)

root.left = Node(2)
root.right = Node(3)

root.left.left = Node(4)

print("Ön Sıralı Geçiş: ", end="")
root.traversePreOrder()
print("\nSıralı Geçiş: ", end="")
root.traverseInOrder()
print("\nArka Sıralı Geçiş: ", end="")
root.traversePostOrder()
```
