# kili Aa 
- kili aa, her bir st dmn en fazla iki ocua sahip olabilecei bir aa veri yapsdr.<br>
  Bir ikili aacn her dm  eden oluur:
  1. Veri esi
  2. Sol ocuun adresi
  3. Sa ocuun adresi<br>
  
  

### kili Aa Trleri:
1. <b>Tam kili Aa:</b><br> Tam kili aa, her st dmn/i dmn iki veya hi ocuu olmayan zel bir ikili aa trdr.<br>
2. <b>Mkemmel kili Aa:</b><br> Mkemmel ikili aa, her dmn tam olarak iki alt dme sahip olduu ve tm yaprak dmlerinin ayn seviyede olduu bir ikili aa trdr.<br>
3. <b>Komple kili Aa:</b><br>Komple ikili aa, tam bir ikili aa gibidir, ancak iki byk fark vardr.
    1. Her seviye tamamen doldurulmaldr.
    2. Tm yaprak elemanlar sola doru eilmelidir.
    3. Son yaprak elemannn bir doru kardei olmayabilir, yani komple bir ikili aacn tam bir ikili aa olmas gerekmez.<br>
4. <b>Dejenere veya Patolojik Aa:</b><br>Dejenere veya patolojik bir aa, sada veya solda tek bir ocuu olan aatr.
5. <b>arpk kili Aa:</b><br>arpk ikili aa, aaca sol dmlerin veya sa dmlerin hakim olduu patolojik/dejenere bir aatr. <br>Bylece, iki tr arpk ikili aa vardr: <b>sola arpk ikili aa</b> ve <b>saa arpk ikili aa.</b>
6. <b>Dengeli kili Aa:</b><br>Her dm iin sol ve sa alt aacn ykseklii arasndaki farkn 0 veya 1 olduu bir ikili aa trdr.


```python
### Pythonda kili Aa lemi ###

class Node:
    def __init__(self, key):  # Aacn deerleri tanmlanr.
        self.left = None
        self.right = None
        self.val = key

        
    def traversePreOrder(self):  # n sral gei fonksiyonu tanmlanr.
        print(self.val, end=' ')
        if self.left:
            self.left.traversePreOrder()
        if self.right:
            self.right.traversePreOrder()


    def traverseInOrder(self):  # Sral gei fonksiyonu tanmlanr.
        if self.left:
            self.left.traverseInOrder()
        print(self.val, end=' ')
        if self.right:
            self.right.traverseInOrder()

            
    def traversePostOrder(self):  # Arka sral gei fonksiyonu tanmlanr.
        if self.left:
            self.left.traversePostOrder()
        if self.right:
            self.right.traversePostOrder()
        print(self.val, end=' ')


root = Node(1)

root.left = Node(2)
root.right = Node(3)

root.left.left = Node(4)

print("n Sral Gei: ", end="")
root.traversePreOrder()
print("\nSral Gei: ", end="")
root.traverseInOrder()
print("\nArka Sral Gei: ", end="")
root.traversePostOrder()
```
