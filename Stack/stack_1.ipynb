# Stack (Yığın)
### Anahtar Kelimeler:
- <b>Stack:</b> Yığın.
- <b>LIFO(<i>Last in First out</i>):</b> Son giren ilk çıkar.

### Tanımı:
- Yığın, son girenin ilk çıktığı doğrusal bir veri yapısıdır.

### Örnek:
- Yığın yapısını üst üste koyulmuş tabaklar gibi düşünebiliriz.
- Yeni bir tabak ekleneceği zaman en üste eklenir. <i>(Last in)</i>
- Bir tabak çıkartılacağı zaman en üstteki tabak çıkartılır. <i>(First out)</i>

### Yığın Üzerindeki Temel İşlemler:
- <b>Push:</b> Bir Yığının en üstüne öğe eklemek için kullanılır.
- <b>Pop:</b> Bir Yığının en üstünden bir öğe çıkarmak için kullanılır.
- <b>IsEmpty:</b> Yığının boş olup olmadığını kontrol eder.
- <b>IsFull:</b> Yığının dolu olup olmadığını kontrol eder.
- <b>Peek:</b> Bir yığının en üstündeki öğenin değerini, o öğeyi çıkarmadan almayı sağlar.

### Yığınların Çalışma Prensibi
6 adımda yığınların çalışma prensibini özetleyebiliriz.

1. Yığının en üst öğesini takip etmek için 'TOP' ismi verilen değişken kullanılır.
2. 'TOP' değişkeni ilk tanımlanırken -1 olarak tanımlanır. Yığın başlatılırken 'TOP == -1' ile karşılaştırılarak yığının boş olup olmadığı kontrol edilir. (İndeks mantığı 0'dan başladığı için TOP değişkenini 0'a değil de -1'e eşitliyoruz başlangıçta.)
3. Bir yığına öğe eklerken 'TOP' değişkeninin değeri 1 arttırılır ve yeni öğe "TOP'un" gösterdiği konuma yerleştirilir.
4. Bir yığından öğe çıkarırken 'TOP' değişkeninin değeri 1 azaltılır ve "TOP'un" konumu da değişir.
5. Yığına öğe eklemeden önce yığının dolu olup olmadığını kontrol edilir.
6. Yığından öğe çıkarmadan önce yığının boş olup olmadığı kontrol edilir.

```python
### Pythonda Yığın İşlemi ###

# Adım 1: Yığını oluşturmak
def create_stack():
    stack = []
    print('Stack oluşturuldu.')
    return stack


# Adım 2: Yığına öğe eklemek
def push(stack, item):
    stack.append(item)
    print('Eklenen Öğe:', item)


# Adım 3: Yığından öğe çıkarmak
def pop(stack):
    if len(stack) == 0:
        return 'Öğe çıkarılamadı. Yığın boş!'
    
    else:
        return stack.pop()
    

stack = create_stack()
push(stack, 1)
push(stack, 2)
push(stack, 3)
push(stack, 4)

print('Çıkarılan Öğe: ' , pop(stack))
print('Öğe Çıkarıldıktan Sonra Yığın: ' , stack)
```
