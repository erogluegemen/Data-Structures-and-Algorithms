# Kuyruk Türleri
<b>Dört çeşit kuyruk türü vardır:</b>
1. Basit Kuyruk (Simple Queue)
2. Dairesel Kuyruk (Circular Queue)
3. Öncelikli Kuyruk (Priority Queue)
4. Çift Uçlu Kuyruk (Double Ended Queue - Deque)

### 1.) Basit Kuyruk:
- Basit kuyrukta, öğe ekleme işlemi arkadan, çıkarma işlemi önden yapılır(FIFO).

### 2.) Dairesel Kuyruk:
- Dairesel kuyrukta, son öğe dairesel bir bağlantı oluşturan ilk öğeyi işaret eder.
- Dairesel kuyruğun, basit kuyruğa göre ana avantajı daha iyi bellek kullanımıdır. Son indeks dolu ve ilk indeks boş ise, ilk indekse bir öğe eklenebilir. Bu işlemin basit kuyrukta yapılması mümkün değil.

### 3.) Öncelikli Kuyruk:
- Öncelikli kuyruk, her öğenin bir öncelik ile ilişkilendirildiği ve önceliğine göre sunulduğu özel bir sıra türüdür.
- Aynı önceliğe sahip öğeler ortaya çıkarsa, sıradaki sıralarına göre servis edilir.

### 4.) Çift Uçlu Kuyruk:
- Çift uçlu bir kuyrukta, öğelerin eklenmesi ve çıkarılması önden veya arkadan gerçekleştirilebilir. <b>Bu nedenle FIFO (İlk Giren İlk Çıkar) kuralına uymaz.</b>
