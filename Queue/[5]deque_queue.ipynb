# ift Ulu Kuyruk
- Deque veya ift ulu kuyruk, elerin eklenmesi ve karlmasnn nden veya arkadan gerekletirilebildii bir kuyruk trdr. Bu nedenle FIFO (lk Giren lk kar) kuralna uymaz.


### ift Ulu Kuyruk Trleri
1. Giri(Input) Kstl:
    * Bu dzenlemede, giri tek bir uta kstlanr, ancak her iki uta da silmeye izin verir.
2. k(Output) Kstl:
    * Bu dzenlemede, kt tek bir uta kstlanr, ancak her iki uta da eklemeye izin verir.
    
### ift Ulu Kuyruun alma Prensibi:
- Dairesel bir dizide, dizi doluysa batan balarz. Ancak bir dorusal dizi uygulamasnda, dizi doluysa, daha fazla e eklenemez.
- Aadaki ilemler yaplmadan nce bu admlar takip edilir.
    1. n boyutunda bir dizi (deque) aln.
    2. ki iaretiyi ilk konuma ayarlayn ve FRONT = -1 ve REAR = 0 olarak tanmla.

1. nden Ekleme:
    - FRONT iaretisinin konumuna bak
    - Eer FRONT < 1 ise, FRONT'u tekrardan FRONT = n-1 olarak tanmla.
    - Delse FRONT'u 1 azalt.
    - Yeni eleman dizine ekle.
    
2. Arkadan Ekleme:
    - Dizin dolu mu diye kontrol edilir.
    - Eer dizin doluysa, REAR' tekrardan REAR = 0 diye tanmla.
    - Deilse REAR'I 1 arttr.
    - Yeni eleman dizine ekle.
    
3. nden Silme:
    - Dizin bo mu diye kontrol edilir.
    - Dizin bo ise(FRONT = -1 ise), silme yaplamaz. (<b>underflow condition</b>)
    - Eer dizinde sadece 1 eleman var ise(FRONT = REAR ise), FRONT = -1, REAR = -1 olarak ayarlanr.
    - Aksi takdirde FRONT sondaysa (FRONT = n - 1 ise), FRONT = 0 olarak ayarlayn.
    - Deilse, FRONT = FRONT + 1.
    
4. Arkadan Silme:
    - Dizin bo mu diye kontrol edilir.
    - Dizin bo ise(FRONT = -1 ise), silme yaplamaz. (<b>underflow condition</b>)
    - Eer dizinde sadece 1 eleman var ise(FRONT = REAR ise), FRONT = -1, REAR = -1 olarak ayarlanr.
    - Aksi takdirde REAR bataysa (REAR = 0 ise), REAR = n-1 olarak ayarlanr.
    - Deilse, REAR = REAR-1.
    
5. Bo mu Dolu mu Kontrol:
    - Eer FRONT = -1 ise botur.
    - Eer FRONT = 0 ve REAR = n-1 veya FRONT = REAR + 1 ise doludur.

```python
### Pythonda ift Ulu Kuyruk lemi ###

class Deque:
    def __init__(self):
        self.items = []  # 

    def isEmpty(self):  # Bo mu diye kontrol edilir.
        return self.items == []

    def addRear(self, item):  # Arkadan e ekleme.
        self.items.append(item)

    def addFront(self, item):  # nden e ekleme.
        self.items.insert(0, item)

    def removeFront(self):  # nden e karma.
        return self.items.pop(0)

    def removeRear(self):  # Arkadan e karma.
        return self.items.pop()

    def size(self):  # Listenin boyutunu yazdrma.
        return len(self.items)


d = Deque()
print(d.isEmpty())     # Liste u anlk bo olduu iin True dndryor.
d.addRear(8)           # eleri ekliyoruz.
d.addRear(5)
d.addFront(7)
d.addFront(10)
print(d.size())        # Listenin boyutu (4)
print(d.isEmpty())     # 4 adet e eklediimiz iin bo mu sorgusu False dndryor.
d.addRear(11)
print(d.removeRear())  # Arkadan e karyoruz.
print(d.removeFront()) # nden e karyoruz.
d.addFront(55)
d.addRear(45)
print(d.items)         # Listenin son halini.
```
